```python
def clean_tweet(tweet):
  '''
  Utility function to clean tweet text by removing links, special characters
  using simple regex statements.
  '''
  return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())

def get_pwords(str):
  count = 0
  a = TextBlob(str)
  for word in a.words:
    b = TextBlob(word)
    if(b.sentiment.polarity>0):
      count = count + 1
  return count

def get_padj(str):
    a = TextBlob(str)
    count = 0
    if(a.sentiment.polarity>0):
      count  = 0
      for word in a.words:
        b= TextBlob(word)
        if(b.tags[0][1]=='JJ' or b.tags[0][1]=='JJR' or b.tags[0][1]=='JJS'):
          count = count + 1
    return count
def get_nadj(str):
    a = TextBlob(str)
    count = 0
    if(a.sentiment.polarity<0):
      count  = 0
      for word in a.words:
        b= TextBlob(word)
        if(b.tags[0][1]=='JJ' or b.tags[0][1]=='JJR' or b.tags[0][1]=='JJS'):
          count = count + 1
    return count
def get_punc(str,pun):
        count = 0
        if pun =='...':
            for j in range(0,len(str)-2):
                if(str[j]=="." and str[j+1]=="." and str[j+2]=="."):
                    count = count + 1
        else:
            for i in range (0, len (str)):
                #Checks whether given character is a punctuation mark
                if str[i] in (pun):
                    count = count + 1
        return count
def is_question(str):
      if get_punc(str,'?') > 0:
            return 1
      else:
            return 0


def is_orginal(tweet):
      try:
          if tweet.retweeted_status:
              return 0
      except:
              return 1
def extract_emojis(str):
        return ''.join(c for c in str if c in emoji.UNICODE_EMOJI)
def get_pemoji(str):
        count = 0
        co = TextBlob(str)
        for i in POSITIVE:
            if i in co:
                count = count +1
        for j in emoticons_happy:
            if i in co:
                count = count +1
        n = extract_emojis(str)
        for k in emopos:
            if k in n:
                count = count +1
        return count
def get_nemoji(str):
        count = 0
        co = TextBlob(str)
        for i in NEGATIVE:
            if i in co:
                count = count +1
        for j in emoticons_sad:
            if i in co:
                count = count +1
        n =extract_emojis(str)
        for k in emoneg:
            if k in n:
                count = count +1
        return count
def get_nwords(str):
        count = 0
        a = TextBlob(str)
        for word in a.words:
            b = TextBlob(word)
            if(b.sentiment.polarity>0):
                count = count + 1
        return count
def get_plevel(str):
        a = TextBlob(str)
        if(a.sentiment.polarity>0):
            for word in a.tags:
                if word[1]=='JJS'or word[1]=='RBS':
                    return 3
                elif word[1]=='JJR'or word[1]=='RBR':
                    return 2
                else:
                    return 1
        else:
            return 0
def get_nlevel(str):
        a = TextBlob(str)
        if(a.sentiment.polarity<0):
            for word in a.tags:
                if word[1]=='JJS'or word[1]=='RBS':
                    return -3
                elif word[1]=='JJR'or word[1]=='RBR':
                    return -2
                else:
                    return -1
        else:
            return 0
def get_padv(str):
        a = TextBlob(str)
        if(a.sentiment.polarity>0):
            for word in a.tags:
                if word[1]=='JJS'or word[1]=='RBS':
                    return 3
                elif word[1]=='JJR'or word[1]=='RBR':
                    return 2
                else:
                    return 1
        else:
            return 0



def get_nadv(str):
        a = TextBlob(str)
        if(a.sentiment.polarity<0):
            for word in a.tags:
                if word[1]=='JJS'or word[1]=='RBS':
                    return -3
                elif word[1]=='JJR'or word[1]=='RBR':
                    return -2
                else:
                    return -1
        else:
            return 0
def get_image(tweet):
        try:
            if tweet.entities['media']:
                 #print(tweet)
                #count += 1
                a =tweet.entities['media']
                b =a[0]['media_url']
                #file = wget.download(b)
                #image = Image.open(file)
                return 1
        except:
            return 0

def get_pnouns(str):
        count = 0
        a=TextBlob(str)
        for word in a.tags:
            if word[1] == 'PRP':
                count = count +1
        return count
def get_style(str):
        b = []
        a= lexicon.analyze(str,categories = categories)
        for word in categories:
            b.append(a[word])
        return b
def get_tweet_sentiment(tweet):
        '''
        Utility function to classify sentiment of passed tweet
        using textblob's sentiment method
        '''
        # create TextBlob object of passed tweet text
        analysis = TextBlob(clean_tweet(tweet))
        # set sentiment
        if analysis.sentiment.polarity > 0:
            return 'positive'
        elif analysis.sentiment.polarity == 0:
            return 'neutral'
        else:
            return 'negative'
def get_sentiment(tweet):
        # create TextBlob object of passed tweet text
        analysis = TextBlob(clean_tweet(tweet))
        # set sentiment
        if analysis.sentiment.polarity > 0:
            return '1'
        elif analysis.sentiment.polarity == 0:
            return '0'
        else:
            return '-1'

def get_punc(str,pun):
        count = 0
        if pun =='...':
            for j in range(0,len(str)-2):
                if(str[j]=="." and str[j+1]=="." and str[j+2]=="."):
                    count = count + 1
        else:
            for i in range (0, len (str)):
                #Checks whether given character is a punctuation mark
                if str[i] in (pun):
                    count = count + 1
        return count



def extract_emojis(str):
  return ''.join(c for c in str if c in emoji.UNICODE_EMOJI)

def get_pemoji(str):
    count = 0
    co = TextBlob(str)
    for i in POSITIVE:
        if i in co:
            count = count +1
    for j in emoticons_happy:
        if i in co:
            count = count +1
    n = extract_emojis(str)
    for k in emopos:
        if k in n:
            count = count +1
    return count
def get_nemoji(str):
    count = 0
    co = TextBlob(str)
    for i in NEGATIVE:
        if i in co:
            count = count +1
    for j in emoticons_sad:
        if i in co:
            count = count +1
    n = extract_emojis(str)
    for k in emoneg:
        if k in n:
            count = count +1
    return count

def get_pwords(str):
    count = 0
    a = TextBlob(str)
    for word in a.words:
        b = TextBlob(word)
        if(b.sentiment.polarity>0):
            count = count + 1
    return count
def get_nwords(str):
    count = 0
    a = TextBlob(str)
    for word in a.words:
        b = TextBlob(word)
        if(b.sentiment.polarity>0):
            count = count + 1
    return count
def get_plevel(str):
    a = TextBlob(str)
    if(a.sentiment.polarity>0):
        for word in a.tags:
            if word[1]=='JJS'or word[1]=='RBS':
                return 3
            elif word[1]=='JJR'or word[1]=='RBR':
                return 2
            else:
                return 1
    else:
        return 0

def get_nlevel(str):
    a = TextBlob(str)
    if(a.sentiment.polarity<0):
        for word in a.tags:
            if word[1]=='JJS'or word[1]=='RBS':
                return -3
            elif word[1]=='JJR'or word[1]=='RBR':
                return -2
            else:
                return -1
    else:
        return 0
def get_padj(str):
    a = TextBlob(str)
    count = 0
    if(a.sentiment.polarity>0):
        count  = 0
        for word in a.words:
            b= TextBlob(word)
            if(b.tags[0][1]=='JJ' or b.tags[0][1]=='JJR' or b.tags[0][1]=='JJS'):
                count = count + 1
    return count


def get_nadj(str):
    a = TextBlob(str)
    count = 0
    if(a.sentiment.polarity<0):
        count  = 0
        for word in a.words:
            b= TextBlob(word)
            if(b.tags[0][1]=='JJ' or b.tags[0][1]=='JJR' or b.tags[0][1]=='JJS'):
                count = count + 1
    return count

def get_padv(str):
    a = TextBlob(str)
    if(a.sentiment.polarity>0):
        for word in a.tags:
            if word[1]=='JJS'or word[1]=='RBS':
                return 3
            elif word[1]=='JJR'or word[1]=='RBR':
                return 2
            else:
                return 1
    else:
        return 0



def get_nadv(str):
    a = TextBlob(str)
    if(a.sentiment.polarity<0):
        for word in a.tags:
            if word[1]=='JJS'or word[1]=='RBS':
                return -3
            elif word[1]=='JJR'or word[1]=='RBR':
                return -2
            else:
                return -1
    else:
        return 0
def get_image(tweet):
    try:
        if tweet.entities['media']:
              #print(tweet)
            #count += 1
            a =tweet.entities['media']
            b =a[0]['media_url']
            #file = wget.download(b)
            #image = Image.open(file)
            return 1
    except:
        return 0
def is_question(str):
    if get_punc(str,'?') > 0:
        return 1
    else:
        return 0
def is_orginal(tweet):
    try:
        if tweet.retweeted_status:
            return 0
    except:
        return 1
def has_link(tweet):
    if tweet.entities['urls']:
        return 1
    else:
        return 0
def get_level(str,result):
    if result == 1:
        k=lexicon.analyze(str, categories=["stress"], normalize=True)
        return (k['stress'])
    elif result == 0:
        k=lexicon.analyze(str, categories=["relaxed"], normalize=True)
        return ((-1)*(k['relaxed']))
def get_pnouns(str):
    count = 0
    a=TextBlob(str)
    for word in a.tags:
        if word[1] == 'PRP':
            count = count +1
    return count
def get_style(str):
    b = []
    a= lexicon.analyze(str,categories = categories)
    for word in categories:
        b.append(a[word])
    return b
def get_sarcasm(str):
    nemoji = get_pemoji(str)
    pemoji = get_nemoji(str)
    nwords = get_nwords(str)
    pwords = get_pwords(str)
    if(pemoji>nemoji and pwords<nwords):
        return 1
    elif(pemoji<nemoji and pwords>nwords):
        return 1
    else:
        return 0






```
